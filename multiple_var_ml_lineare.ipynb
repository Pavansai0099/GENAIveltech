```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/home price multiple var.csv')
df
```

```python
import math
mediam=math.floor(df.bedroom.median())
mediam
```

```python
df.bedroom=df.bedroom.fillna(mediam)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
