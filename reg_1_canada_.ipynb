```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/canada_per_capita_income 1.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df.price , color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.price)
```

```python
reg.predict([[2017]])
```

```python
reg.intercept_
```

```python
828.46507522*2017+-1632210.7578554575
```

```python
d = pd.read_csv('/content/reg1.csv')
d.head(5)
```

```python
p=reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_csv("reg1.csv")
```
